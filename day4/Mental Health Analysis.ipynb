# DATA VISUALIZATION

```python
#a basic sample for visualization that showing Average Age by Gender
avg_age = df.groupby('Gender')['Age'].mean()

plt.figure(figsize=(6, 4))
avg_age.plot(kind='bar', color=['skyblue', 'salmon'])
plt.title("Average Age by Gender")
plt.xlabel('Gender')
plt.ylabel('Age')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
#methods
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from xgboost import XGBClassifier


from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
```

```python
#dropped unimportant columns
df=df.drop(columns=["Employee_ID","Name"])
```

```python
#select x
x=df.drop(["Mental_Health_Status"],axis=1)
```

```python
df.info()
```

```python
for temp in x.columns:
    if x[temp].dtype == object:
        encoder = LabelEncoder()
        x[temp] = encoder.fit_transform(x[temp])

if df["Mental_Health_Status"].dtype == object:
    encoder = LabelEncoder()
    df["Mental_Health_Status"] = encoder.fit_transform(df["Mental_Health_Status"])

y = df["Mental_Health_Status"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=44)
```

```python
methods = [
    LogisticRegression(solver="saga"),
    DecisionTreeClassifier(),
    RandomForestClassifier(),
    GradientBoostingClassifier(),
    KNeighborsClassifier(),
    SVC(),
    XGBClassifier(eval_metric='mlogloss')
]
```

```python
#a function to evaluate all models in one execute
def evaluateModel():
    for model in methods:
        model.fit(x_train, y_train)
        preds = model.predict(x_test)
        acc = accuracy_score(y_test, preds)
        print(f"{model.__class__.__name__} accuracy is: {acc:6f}")
```

```python
evaluateModel()
```

# Model Accucary Visualization


```python
model_names = [
    "LogisticRegression",
    "DecisionTreeClassifier",
    "RandomForestClassifier",
    "GradientBoostingClassifier",
    "KNeighborsClassifier",
    "SVC",
    "XGBClassifier"
]

accuracies = [
    0.40,
    0.25,
    0.30,
    0.30,
    0.35,
    0.35,
    0.50
]

plt.figure(figsize=(10, 6))
bars = plt.bar(model_names, accuracies)

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2.0, height, f"{height:.2f}", ha='center', va='bottom')

plt.xticks(rotation=45)
plt.ylabel("Accuracy")
plt.title("Model Accuracy Comparison")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
