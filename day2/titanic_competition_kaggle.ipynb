```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.impute import SimpleImputer
```

```python
train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")
```

```python
train.head(5)
```

```python
#finding nan counts 
train.isnull().sum()
```

```python
#replaced Age column's NAN values with mean()
imputer=SimpleImputer(strategy="mean")
train["Age"]=imputer.fit_transform(train[["Age"]])
```

```python
train.isnull().sum()
```

```python
train.head(3)
```

```python
train=pd.get_dummies(train,columns=["Sex"])
train.head(3)
```

```python
#embarked column
```

```python
train['Embarked'].fillna(train['Embarked'].mode()[0], inplace=True)
train = pd.get_dummies(train, columns=['Embarked'], drop_first=True)
```

```python
train.head(3)
```

```python
train.isnull().sum()
```

```python
y_train=train["Survived"]
x_train=train.drop(columns=["Survived","Cabin","Name","Ticket"])
test["Survived"]=np.nan
y_test=test["Survived"]
x_test=test.drop(columns=["Survived","Cabin","Name","Ticket"])
```

```python
model=RandomForestClassifier()
model.fit(x_train,y_train)
```

```python
# get_dummies
x_train = pd.get_dummies(x_train)
x_test = pd.get_dummies(x_test)

# eksik olan sütunları tamamla
x_test = x_test.reindex(columns=x_train.columns, fill_value=0)
```

```python
test["Survived"]=model.predict(x_test)
```

```python
df=pd.DataFrame()
df["PassengerId"]=test["PassengerId"]
df["Survived"]=test["Survived"]
```

```python
df.to_csv("Submission.csv",index=False)
```
