```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df=pd.read_excel("Coffe_sales.xlsx")
```

```python
#maybe we can drop "hours_of_day" too, because it's also included with "Time_of_Day"
df=df.drop(["date","datetime","card","Weekday","Month_name"],axis=1)
```

```python
#df=pd.get_dummies(df,columns=["cash_type","coffee_name","Time_of_Day"])
columns=["cash_type","coffee_name","Time_of_Day"]
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
for x in columns:   
    df[x] = le.fit_transform(df[x])
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split

y=df["coffee_name"]
x=df.drop(columns=["coffee_name"])

x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7,random_state=31)
```

```python
x_train=x_train.values if hasattr(x_train,"values") else x_train
x_test=x_test.values if hasattr(x_test,"values") else x_test
```

```python
from sklearn.ensemble import RandomForestClassifier

classModel=RandomForestClassifier(max_depth=6)
classModel.fit(x_train,y_train)
```

```python
classModel.predict(x_test)
classModel.score(x_test,y_test)
```
